```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
url = 'https://en.wikipedia.org/wiki/Wikipedia:10,000_most_common_passwords'
page_response = requests.get(url)
soup = BeautifulSoup(page_response.text, 'html')
```

```python
print(soup.prettify())
```

```python
most100_common_passwords = soup.find_all('ol')[0] #The first 100
```

```python
most1000_common_passwords = (soup.find_all('ol')[1]) #The rest of the 101-1000
```

```python
most100_common_passwords_list_raw = [str(password.text.strip()) for password in most100_common_passwords] #Every element had a '' after
most1000_common_passwords_list_raw = [str(password.text.strip()) for password in most1000_common_passwords] #Every element had a '' after
```

```python
most100_common_passwords_list = [element for element in most100_common_passwords_list_raw if element not in ('')] # Remove the ''
most1000_common_passwords_list = [element for element in most1000_common_passwords_list_raw if element not in ('')] #Remove the ''
```

```python
most_common_passwords_list = most100_common_passwords_list + most1000_common_passwords_list
```

```python
df = pd.DataFrame({'Most Common Passwords' : most_common_passwords_list})
```
